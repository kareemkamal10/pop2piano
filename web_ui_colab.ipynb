## üìå Quick Access (Alternative Methods)

```python
#@title Option B: Use ngrok for permanent URL
#@markdown Get your auth token from https://dashboard.ngrok.com/

use_ngrok = False #@param {type:"boolean"}
ngrok_token = "" #@param {type:"string"}

if use_ngrok and ngrok_token:
    !pip install -q pyngrok
    from pyngrok import ngrok
    
    ngrok.set_auth_token(ngrok_token)
    public_url = ngrok.connect(7860)
    print(f"üåê Public URL: {public_url}")
    
    # Launch without Gradio's share
    app.launch(share=False, server_port=7860)
```

```python
#@title Option C: Use Cloudflare Tunnel

use_cloudflare = False #@param {type:"boolean"}

if use_cloudflare:
    !wget -q https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64 -O cloudflared
    !chmod +x cloudflared
    
    import subprocess
    import threading
    
    def run_cloudflare():
        subprocess.run(['./cloudflared', 'tunnel', '--url', 'http://localhost:7860'])
    
    thread = threading.Thread(target=run_cloudflare)
    thread.start()
    
    # Launch the app
    app.launch(share=False, server_port=7860)
```
